```python
df = data.drop(["cityCode","basement","attic","garage"],axis=1)
df.head()
```

```python
sns.scatterplot(x=df["squareMeters"],y=df["price"])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,mean_squared_error
from math import sqrt
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, Activation, Dropout
from tensorflow.keras.optimizers import Adam
```

```python
x = df.drop(["price"],axis=1)
y = df["price"]
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled = scaler.fit_transform(x)
x = pd.DataFrame(scaled,index=x.index,columns=x.columns)
```

```python
x.head()
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=101)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Linear Regression

```python
lin_reg = LinearRegression()
lin_reg.fit(x_train,y_train)
lin_pred = lin_reg.predict(x_test)
print("Mean Squared Error : {}".format(mean_squared_error(y_test,lin_pred)))
print("Root Mean Squared Error : {}".format(sqrt(mean_squared_error(y_test,lin_pred))))
print("R-Square : {}".format(r2_score(y_test,lin_pred)))
```

## Ridge Regression

```python
rid_reg = Ridge()
rid_reg.fit(x_train,y_train)
rid_pred = rid_reg.predict(x_test)
print("Mean Squared Error : {}".format(mean_squared_error(y_test,rid_pred)))
print("Root Mean Squared Error : {}".format(sqrt(mean_squared_error(y_test,rid_pred))))
print("R-Square : {}".format(r2_score(y_test,rid_pred)))
```

## Lasso Regression

```python
las_reg = Lasso()
las_reg.fit(x_train,y_train)
las_pred = las_reg.predict(x_test)
print("Mean Squared Error : {}".format(mean_squared_error(y_test,las_pred)))
print("Root Mean Squared Error : {}".format(sqrt(mean_squared_error(y_test,las_pred))))
print("R-Square : {}".format(r2_score(y_test,las_pred)))
```

## Elastic Net Regression

```python
ela_reg = ElasticNet()
ela_reg.fit(x_train,y_train)
ela_pred = ela_reg.predict(x_test)
print("Mean Squared Error : {}".format(mean_squared_error(y_test,ela_pred)))
print("Root Mean Squared Error : {}".format(sqrt(mean_squared_error(y_test,ela_pred))))
print("R-Square : {}".format(r2_score(y_test,ela_pred)))
```

## KNeighborsRegressor

```python
knn_reg = KNeighborsRegressor(n_neighbors=7)
knn_reg.fit(x_train,y_train)
knn_pred = knn_reg.predict(x_test)
print("Mean Squared Error : {}".format(mean_squared_error(y_test,knn_pred)))
print("Root Mean Squared Error : {}".format(sqrt(mean_squared_error(y_test,knn_pred))))
print("R-Square : {}".format(r2_score(y_test,knn_pred)))
```

## Support Vector Regressor

```python
sv_reg = SVR()
sv_reg.fit(x_train,y_train)
sv_reg_pred = sv_reg.predict(x_test)
print("Mean Squared Error : {}".format(mean_squared_error(y_test,sv_reg_pred)))
print("Root Mean Squared Error : {}".format(sqrt(mean_squared_error(y_test,sv_reg_pred))))
print("R-Square : {}".format(r2_score(y_test,sv_reg_pred)))
```

## Random Forest Regressor

```python
rf_reg = RandomForestRegressor()
rf_reg.fit(x_train,y_train)
rf_reg_pred = rf_reg.predict(x_test)
print("Mean Squared Error : {}".format(mean_squared_error(y_test,rf_reg_pred)))
print("Root Mean Squared Error : {}".format(sqrt(mean_squared_error(y_test,rf_reg_pred))))
print("R-Square : {}".format(r2_score(y_test,rf_reg_pred)))
```

## Artificial Neural Networks

```python
model =  Sequential()

model.add(Dense(13,activation="relu"))
model.add(Dense(32,activation="relu"))
model.add(Dropout(0.25))

model.add(Dense(64,activation="relu"))
model.add(Dropout(0.25))

model.add(Dense(128,activation="relu"))
model.add(Dropout(0.25))

model.add(Dense(512,activation="relu"))
model.add(Dropout(0.25))
model.add(Dense(1))

model.compile(optimizer="adam",loss="mse")

ann_model = model.fit(x_train,y_train,
                     validation_data=(x_test,y_test),
                     epochs=200)
```

```python
pd.DataFrame(ann_model.history).plot()
```

```python
ann_pred = model.predict(x_test)
```

```python
print("Mean Squared Error : {}".format(mean_squared_error(y_test,ann_pred)))
print("Root Mean Squared Error : {}".format(sqrt(mean_squared_error(y_test,ann_pred))))
print("R-Square : {}".format(r2_score(y_test,ann_pred)))
```

## Evaluate Models

```python
mse = [mean_squared_error(y_test,lin_pred),
      mean_squared_error(y_test,rid_pred),
      mean_squared_error(y_test,las_pred),
      mean_squared_error(y_test,ela_pred),
      mean_squared_error(y_test,knn_pred),
      mean_squared_error(y_test,sv_reg_pred),
      mean_squared_error(y_test,rf_reg_pred),
      mean_squared_error(y_test,ann_pred)]


rmse = [sqrt(mean_squared_error(y_test,lin_pred)),
       sqrt(mean_squared_error(y_test,rid_pred)),
       sqrt(mean_squared_error(y_test,las_pred)),
       sqrt(mean_squared_error(y_test,ela_pred)),
       sqrt(mean_squared_error(y_test,knn_pred)),
       sqrt(mean_squared_error(y_test,sv_reg_pred)),
       sqrt(mean_squared_error(y_test,rf_reg_pred)),
       sqrt(mean_squared_error(y_test,ann_pred))]


r2 = [r2_score(y_test,lin_pred),
     r2_score(y_test,rid_pred),
     r2_score(y_test,las_pred),
     r2_score(y_test,ela_pred),
     r2_score(y_test,knn_pred),
     r2_score(y_test,sv_reg_pred),
     r2_score(y_test,rf_reg_pred),
     r2_score(y_test,ann_pred)]


mse_df = []
for i in mse:
    mse_df.append(i)
    
    
    
rmse_df = []
for i in rmse:
    rmse_df.append(i)
    
    
r2_df = []
for i in r2:
    r2_df.append(i)
```

```python
eva_df = pd.DataFrame([mse_df,rmse_df,r2_df],index=["Mean Squared Error","Root Mean Squared Error","R-Square"],
                     columns=["Linear Regression",
                             "Ridge Regression",
                             "Lasso Regression",
                             "Elastic Net Regression",
                             "KNeighborsRegressor",
                             "Support Vector Regressor",
                             "Random Forest Regressor",
                             "Artificial Neural Networks"]).T
```

```python
plt.figure(figsize=(15,15))
plt.subplot(2,2,1)
eva_df["R-Square"].plot(kind="barh",color="orange")
plt.subplot(2,2,1)
eva_df[["R-Square","Root Mean Squared Error"]]
```
